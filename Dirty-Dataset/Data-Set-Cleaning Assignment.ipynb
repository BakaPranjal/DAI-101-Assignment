# NUMERICAL ANALYSIS

```python
column_name="ApplicantIncome"
Q1 = df[column_name].quantile(0.25)
Q3 = df[column_name].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df[column_name] < (Q1 - 1.5 * IQR)) | (df[column_name] > (Q3 + 1.5 * IQR))]
plt.boxplot(df[column_name])
plt.show()
plt.scatter(x=outliers.index, y=outliers[column_name], color='red')
plt.show()
```

```python
Q1 = df["ApplicantIncome"].quantile(0.25)
Q3 = df["ApplicantIncome"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df = df[(df['ApplicantIncome'] >= lower) & (df['ApplicantIncome'] <= upper)]
```

# BI-VARIATE ANALYSIS

### Applicant Income vs Loan Status:

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Loan_Status', y='ApplicantIncome')
plt.title('Applicant Income vs Loan Status')
plt.show()
```

There is a wide variation in income for both loan approval statuses, though it appears that applicants with higher incomes are not necessarily more likely to get approved.

### Credit History vs Loan Status:

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Credit_History', hue='Loan_Status')
plt.title('Credit History vs Loan Status')
plt.show()
```

Applicants with a credit history of 1 (positive credit history) are much more likely to get their loans approved.

### Education vs Applicant Income:

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Education', y='ApplicantIncome')
plt.title('Education vs Applicant Income')
plt.show()
```

Graduates tend to have higher incomes compared to non-graduates, but there is a large overlap.
